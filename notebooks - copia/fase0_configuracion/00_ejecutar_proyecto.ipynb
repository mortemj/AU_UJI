```python
# ============================================================================
# FUNCIONES DE VISUALIZACIÓN
# ============================================================================

def barra_progreso_notebooks(notebooks_info, actual_idx=-1, estado='pendiente'):
    """
    Genera HTML de barra de progreso para notebooks.
    
    notebooks_info: lista de (nombre, completado, tiempo)
    actual_idx: índice del notebook actual (-1 si ninguno)
    estado: 'ejecutando', 'completado', 'error'
    """
    html = '<div style="margin: 20px 0;">'
    
    for i, (nombre, completado, tiempo) in enumerate(notebooks_info):
        # Determinar estado visual
        if completado == True:
            bg = '#38a169'
            icono = '✅'
            texto = f'{tiempo:.1f}s'
        elif completado == False:
            bg = '#e53e3e'
            icono = '❌'
            texto = f'{tiempo:.1f}s'
        elif i == actual_idx:
            bg = '#3182ce'
            icono = '🔄'
            texto = 'ejecutando...'
        else:
            bg = '#e2e8f0'
            icono = '⏳'
            texto = 'pendiente'
        
        color_texto = 'white' if bg not in ['#e2e8f0'] else '#718096'
        
        html += f'''
        <div style="display: flex; align-items: center; padding: 10px 15px; 
                    background: {bg}; color: {color_texto}; margin: 5px 0; 
                    border-radius: 8px;">
            <span style="font-size: 18px; margin-right: 12px;">{icono}</span>
            <span style="flex: 1; font-weight: 500;">{nombre}</span>
            <span style="font-size: 14px;">{texto}</span>
        </div>
        '''
    
    html += '</div>'
    return html


def resumen_final(resultados, tiempo_total):
    """Genera HTML del resumen final."""
    exitos = sum(1 for _, ok, _ in resultados if ok)
    total = len(resultados)
    
    if exitos == total:
        color_bg, icono, mensaje = '#38a169', '🎉', '¡EJECUCIÓN COMPLETADA!'
    else:
        color_bg, icono, mensaje = '#e53e3e', '⚠️', f'COMPLETADOS: {exitos}/{total}'
    
    html = f'''
    <div style="background: linear-gradient(135deg, {color_bg} 0%, {color_bg}dd 100%); 
                padding: 30px; border-radius: 15px; margin: 20px 0;">
        <h2 style="color: white; margin: 0 0 20px 0; text-align: center;">{icono} {mensaje}</h2>
        <div style="display: flex; justify-content: center; gap: 40px;">
            <div style="text-align: center;">
                <div style="color: white; font-size: 36px; font-weight: bold;">{exitos}/{total}</div>
                <div style="color: rgba(255,255,255,0.8);">Notebooks OK</div>
            </div>
            <div style="text-align: center;">
                <div style="color: white; font-size: 36px; font-weight: bold;">{tiempo_total:.1f}s</div>
                <div style="color: rgba(255,255,255,0.8);">Tiempo total</div>
            </div>
        </div>
    </div>
    '''
    return html

print('✅ Funciones de visualización cargadas')
```

## ▶️ Ejecutar notebooks

```python
# ============================================================================
# EJECUTAR NOTEBOOKS CON PROGRESO VISUAL
# ============================================================================

import nbformat
from nbconvert.preprocessors import ExecutePreprocessor

def ejecutar_notebook(ruta, timeout=600):
    """Ejecuta un notebook y devuelve (exito, tiempo, error)."""
    inicio = time.time()
    try:
        with open(ruta, 'r', encoding='utf-8') as f:
            nb = nbformat.read(f, as_version=4)
        
        ep = ExecutePreprocessor(timeout=timeout, kernel_name='python3')
        ep.preprocess(nb, {'metadata': {'path': str(ruta.parent)}})
        
        with open(ruta, 'w', encoding='utf-8') as f:
            nbformat.write(nb, f)
        
        return True, time.time() - inicio, None
    except Exception as e:
        return False, time.time() - inicio, str(e)[:200]

# Inicializar estado
# None = pendiente, True = ok, False = error
estado_notebooks = [[nb[0], None, 0] for nb in notebooks_ejecutar]
tiempo_inicio_total = time.time()

# Título
titulo_html = '''
<div style="background: linear-gradient(135deg, #2c5282 0%, #1a365d 100%); 
            padding: 20px; border-radius: 10px; margin: 20px 0;">
    <h2 style="color: white; margin: 0; text-align: center;">🚀 EJECUTANDO NOTEBOOKS</h2>
</div>
'''

# Ejecutar cada notebook
for i, (nombre, ruta) in enumerate(notebooks_ejecutar):
    # Actualizar display
    clear_output(wait=True)
    display(HTML(banner_html))
    display(HTML(titulo_html))
    display(HTML(barra_progreso_notebooks(estado_notebooks, actual_idx=i)))
    
    # Ejecutar
    exito, tiempo, error = ejecutar_notebook(ruta)
    
    # Actualizar estado
    estado_notebooks[i] = [nombre, exito, tiempo]
    
    # Log de error si falló
    if not exito:
        print(f"\n❌ Error en {nombre}:")
        print(f"   {error}")

tiempo_total = time.time() - tiempo_inicio_total

# Mostrar resultado final
clear_output(wait=True)
display(HTML(banner_html))
display(HTML(resumen_final(estado_notebooks, tiempo_total)))
display(HTML('<h3 style="margin: 20px 0 10px;">📋 Detalle:</h3>'))
display(HTML(barra_progreso_notebooks(estado_notebooks)))
```
