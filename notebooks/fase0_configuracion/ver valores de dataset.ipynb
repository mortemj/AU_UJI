```python
import pandas as pd
import joblib
from pathlib import Path

ROOT = Path(r"C:\Users\mjmor\OneDrive - Universitat Jaume I\2.- AU_UJI")

# Dataset
df = pd.read_parquet(ROOT / "data" / "automl" / "dataset_final_tfm.parquet")
print("=== Columnas ===")
print(df.columns.tolist())
print("\n=== situacion_laboral ===")
print(df["situacion_laboral"].value_counts())
print("\n=== tipo_acceso ===")
print(df["tipo_acceso"].value_counts())

# Pipeline
pipeline = joblib.load(ROOT / "data" / "05_modelado" / "pipeline_preprocesamiento.pkl")
print("\n=== feature_names_in_ ===")
print(list(pipeline.feature_names_in_) if hasattr(pipeline, "feature_names_in_") else "NO TIENE")
print("\n=== steps ===")
for name, step in pipeline.steps:
    print(f"  {name}: {type(step).__name__}")
```
