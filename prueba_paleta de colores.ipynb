```python
import matplotlib.pyplot as plt

def save_palette_as_png(colors, filename="palette.png", title="Palette"):
    plt.figure(figsize=(12, 2))
    for i, color in enumerate(colors):
        plt.bar(i, 1, color=color)
    plt.xticks(range(len(colors)), colors, rotation=45, ha="right")
    plt.yticks([])
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename, dpi=200)
    plt.close()

# Guardar las 3 paletas
save_palette_as_png(palette_modern_pastel_science, "palette_main.png", "Paleta Principal")
save_palette_as_png(palette_dark_mode, "palette_dark.png", "Paleta Fondo Oscuro")
save_palette_as_png(palette_accents, "palette_accents.png", "Paleta de Acentos")
```

```python
def save_palette_as_png(colors, filename="palette.png", title="Palette", background="#ffffff"):
    plt.figure(figsize=(12, 2), facecolor=background)
    ax = plt.gca()
    ax.set_facecolor(background)

    for i, color in enumerate(colors):
        plt.bar(i, 1, color=color)

    text_color = "white" if background != "#ffffff" else "black"

    plt.xticks(range(len(colors)), colors, rotation=45, ha="right", color=text_color)
    plt.yticks([], color=text_color)
    plt.title(title, color=text_color)
    plt.tight_layout()
    plt.savefig(filename, dpi=200, facecolor=background)
    plt.close()
```

```python
save_palette_as_png(palette_modern_pastel_science, "palette_main.png", "Paleta Principal", background="#ffffff")
save_palette_as_png(palette_dark_mode, "palette_dark.png", "Paleta Fondo Oscuro", background="#1e1e1e")
save_palette_as_png(palette_accents, "palette_accents.png", "Paleta de Acentos", background="#ffffff")
```
