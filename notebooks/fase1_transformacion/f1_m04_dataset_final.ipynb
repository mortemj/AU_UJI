## 🎯 ¿Qué hace este notebook?

Es un **orquestador** que ejecuta secuencialmente:

1. **M04a**: Une 8 tablas del Excel principal → `df_alumno_base.parquet`
2. **M04b**: Añade preinscripción + genera HTML → `df_alumno.parquet`
3. **M04c**: Corrige notas de acceso (normalización + cascada) → sobreescribe `df_alumno.parquet`
4. **M04d**: Corrige vía de acceso (duplicados + renombre) → sobreescribe `df_alumno.parquet`

Muestra progreso visual con banners y barra de tiempo.

## ⚠️ Requisitos

- Haber ejecutado `f1_m02_limpieza.ipynb` (genera los parquets en `01_interim/`)
- El módulo `src/display.py` con las funciones de presentación visual

## 📦 ¿Qué genera?

| Archivo | Ubicación | Generado por |
|---|---|---|
| df_alumno_base.parquet | `data/02_processed/` | M04a |
| df_alumno.parquet | `data/02_processed/` | M04b → M04c → M04d |
| df_alumno.csv | `data/02_processed/` | M04d |
| m04_dataset_final.html | `docs/html/fase1/` | M04b |

```python
# ============================================================================
# CELDA 12: RESUMEN FINAL
# ============================================================================

import pandas as pd
from src.config import RUTA_PROCESSED
from src.utils import formato_numero_es
fmt = formato_numero_es

path_final = RUTA_PROCESSED / 'df_alumno.parquet'
if path_final.exists():
    _df = pd.read_parquet(path_final)
    _n_reg = len(_df)
    _n_col = len(_df.columns)
    _n_alu = _df['per_id_ficticio'].nunique()
    _size = path_final.stat().st_size / 1024 / 1024
else:
    _n_reg = _n_col = _n_alu = 0
    _size = 0

tiempo_total = tiempo_a + tiempo_b + tiempo_c + tiempo_d

STATS = [
    {'valor': '9',                   'titulo': 'Tablas unidas',  'icono': '📊'},
    {'valor': fmt(_n_reg),           'titulo': 'Registros',      'icono': '📋'},
    {'valor': str(_n_col),           'titulo': 'Columnas',       'icono': '📐'},
    {'valor': f'{tiempo_total:.1f}s','titulo': 'Tiempo total',   'icono': '⏱️'},
]

mostrar_resumen_final(
    titulo='¡M04 COMPLETADO CON ÉXITO!',
    stats=STATS,
    icono='🎉',
    color='verde'
)
```

```python
# ============================================================================
# CELDA 9: ARCHIVOS GENERADOS
# ============================================================================
#
# Muestra los archivos generados con sus tamaños reales.
# Calcula dinámicamente el tamaño de cada archivo.
# ============================================================================

from IPython.display import HTML, display

# Calcular tamaños dinámicamente
archivos = [
    ('df_alumno.parquet',      RUTA_PROCESSED, '🐍', 'Python / Análisis',          '#3182ce', '#ebf8ff'),
    ('df_alumno.csv',          RUTA_PROCESSED, '📊', 'Tableau / Power BI',          '#38a169', '#f0fff4'),
    ('df_alumno_base.parquet', RUTA_PROCESSED, '📦', 'Intermedio (sin preinscripción)', '#805ad5', '#faf5ff'),
]

items_html = ''
for i, (nombre, ruta, emoji, uso, color, bg) in enumerate(archivos):
    path = ruta / nombre
    if path.exists():
        size = f'{path.stat().st_size / 1024 / 1024:.2f} MB'
    else:
        size = 'No generado'
    items_html += f'''
    <div class="archivo-item" style="background:{bg}; border-left:4px solid {color};
         padding:15px; margin:8px 0; border-radius:10px; display:flex; align-items:center;
         animation:slideIn 0.5s ease-out {0.1+i*0.1}s forwards; opacity:0;">
        <span style="font-size:1.5em; margin-right:15px;">{emoji}</span>
        <div>
            <strong style="color:{color};">{nombre}</strong>
            <p style="margin:3px 0 0 0; color:#718096; font-size:0.85em;">{uso} ({size})</p>
        </div>
    </div>'''

html = f'''
<style>
@keyframes slideIn {{
    from {{ opacity: 0; transform: translateX(-20px); }}
    to {{ opacity: 1; transform: translateX(0); }}
}}
</style>
<div style="background:white; border-radius:12px; padding:20px; margin:15px 0;
     box-shadow:0 4px 15px rgba(0,0,0,0.1);">
    <h4 style="margin:0 0 15px 0; color:#2d3748;">📁 Archivos Generados</h4>
    {items_html}
    <p style="margin-top:15px; color:#718096; font-size:0.85em; text-align:center;">
        📍 Ubicación: <code>{RUTA_PROCESSED}</code>
    </p>
</div>
'''
display(HTML(html))
```
