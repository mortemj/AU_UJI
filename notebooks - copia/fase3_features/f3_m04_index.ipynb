## 🎯 Objetivo

Este modulo M04 presenta DOS versiones diferentes de encoding de variables categoricas:

1. **M04 AutoML**: One-hot encoding COMPLETO (100% numerico)
   - Columnas: DINÁMICAS (leer celda siguiente)
   - Compatible con PyCaret, H2O, AutoGluon
   - Output: df_exp_automl_target.parquet

2. **M04 EDA**: Encoding PARCIAL (mezcla numerico + strings)
   - Columnas: DINÁMICAS (leer celda siguiente)
   - Compatible con CatBoost, LightGBM, XGBoost
   - Output: df_exp_target_eda.parquet

---

## 📊 Comparativa

| Aspecto | AutoML | EDA |
|---------|--------|-----|
| Columnas | DINÁMICAS | DINÁMICAS |
| Tipo | 100% Numérico | Mixto (strings + números) |
| Rama | rama (codificada 1-5) | rama (string: TE, HU, ...) |
| Provincia | provincia (codificada 0-5) | provincia (string: Castelló, ...) |
| Sexo | 0/1 (numerico) | 'Hombre'/'Mujer' (string) |
| Via Acceso | via_acceso (codificada 0-13) | via_acceso (string) |
| Uso recomendado | Fase 3.5 (AutoML) | Fase 4 (EDA + Modelado) |
