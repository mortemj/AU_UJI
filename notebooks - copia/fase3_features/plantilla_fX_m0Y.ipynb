```python
# ============================================================================
# MÉTRICAS M0Y
# ============================================================================
totales = df.shape[0]
columnas = df.shape[1]
nulos = df.isna().sum().sum()

tablas_info = [
    {"nombre": "dataset", "filas": totales, "columnas": columnas, "nulos": nulos}
]

print("✅ Métricas calculadas")
```

2. Fragmentos dinámicos
Generar botones y paneles HTML a partir de tablas_info.

```python
# ============================================================================
# FRAGMENTOS HTML (botones y paneles)
# ============================================================================
from src.html import generar_html_botones_paneles

botones_html, paneles_html = generar_html_botones_paneles(tablas_info)

print("✅ Fragmentos HTML generados")
```

3. Navegación
Construir navegación dinámica de fases y módulos.

```python
# ============================================================================
# NAVEGACIÓN HTML (fases y módulos)
# ============================================================================
from src.html import generar_html_navegacion_completa

nav_fases_html, nav_modulos_html = generar_html_navegacion_completa(
    fase_activa="faseX",   # ej. "fase1"
    modulo_activo="m0Y"    # ej. "m02"
)

print("✅ Navegación HTML generada")
```

4. Renderizar plantilla con Jinja2
Usar base.html para generar el HTML final.

```python
# ============================================================================
# GENERAR HTML M0Y (usando Jinja2 y base.html)
# ============================================================================
from jinja2 import Environment, FileSystemLoader
from datetime import datetime

env = Environment(loader=FileSystemLoader("src/html/templates"))
template = env.get_template("base.html")

html_completo = template.render(
    titulo="Nombre del módulo",
    subtitulo=f"{len(tablas_info)} tablas procesadas",
    icono="📊",  # cambia según módulo
    nav_fases=nav_fases_html,
    nav_modulos=nav_modulos_html,
    contenido=f"""
        <div class="tablas-grid">{botones_html}</div>
        {paneles_html}
    """,
    autora=AUTORA,
    email_uoc="mjmorteruiz@uoc.edu",
    email_uji=EMAIL_UJI,
    ruta_assets="assets",
    fecha_generacion=datetime.now().strftime("%Y-%m-%d %H:%M"),
    estilos_adicionales="",
    scripts_adicionales=""
)

ruta_html = RUTA_FASEX / "m0Y_nombre.html"
with open(ruta_html, "w", encoding="utf-8") as f:
    f.write(html_completo)

print(f"✅ HTML generado: {ruta_html}")
```
