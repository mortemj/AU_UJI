````python
# ============================================================================
# CELDA 1: CONFIGURACIÓN
# ============================================================================

import time
from datetime import datetime
from pathlib import Path

# Notebooks a ejecutar en orden
NOTEBOOKS = [
    'f5_m01a_preparacion.ipynb',
    'f5_m01b_lineales_basico.ipynb',
    'f5_m01c_lineales_ext.ipynb',
    'f5_m01d_lineales.ipynb',
]

print('=' * 55)
print('MAESTRO M01 — Familia Lineal')
print('=' * 55)
print(f'Inicio: {datetime.now().strftime("%d/%m/%Y %H:%M:%S")}')
print(f'Notebooks: {len(NOTEBOOKS)}')
print()
for nb in NOTEBOOKS:
    existe = Path(nb).exists()
    print(f'  {"✅" if existe else "❌"} {nb}')
````

````python
# ============================================================================
# CELDA 2: EJECUCIÓN SECUENCIAL
# ============================================================================

resultados = {}
t_total_ini = time.time()

for nb in NOTEBOOKS:
    if not Path(nb).exists():
        print(f'❌ {nb} no encontrado — saltando')
        resultados[nb] = ('❌', 0, 'fichero no encontrado')
        continue

    print(f'\n▶️  {nb}')
    print(f'   Inicio: {datetime.now().strftime("%H:%M:%S")}')
    t0 = time.time()

    try:
        %run {nb}
        t = time.time() - t0
        resultados[nb] = ('✅', round(t/60, 1), '')
        print(f'   ✅ Completado en {t/60:.1f} min')
    except Exception as e:
        t = time.time() - t0
        resultados[nb] = ('❌', round(t/60, 1), str(e)[:80])
        print(f'   ❌ Falló en {t/60:.1f} min: {e}')
        print(f'   ⚠️  Continuando con el siguiente...')

t_total = time.time() - t_total_ini
print()
print('=' * 55)
print('RESUMEN FINAL')
print('=' * 55)
for nb, (estado, mins, error) in resultados.items():
    nombre_corto = nb.replace('.ipynb','').replace('f5_','')
    print(f'  {estado}  {nombre_corto:<30} {mins:>5.1f} min  {error}')
print(f'\n  Total: {t_total/60:.1f} min')
print(f'  Fin:   {datetime.now().strftime("%H:%M:%S")}')

n_ok = sum(1 for v in resultados.values() if v[0] == '✅')
print(f'\n  {n_ok}/{len(NOTEBOOKS)} notebooks completados')
if n_ok == len(NOTEBOOKS):
    print('  ✅ Familia lineal completa — listo para f5_m02_arboles')
else:
    print('  ⚠️  Revisa los errores antes de continuar con M02')
````
