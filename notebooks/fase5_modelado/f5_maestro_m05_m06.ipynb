# Notebook Maestro — Fase 5: M05 + M06

Ejecuta `f5_m05_mlp_ebm.ipynb` y `f5_m06_ensambles.ipynb` secuencialmente.
Si uno falla, el otro continúa. Resultado al final.


```python
import time
from datetime import datetime

notebooks = [
    'f5_m05_mlp_ebm.ipynb',
    'f5_m06_ensambles.ipynb',
]

resultados = {}
print(f'🚀 Inicio: {datetime.now().strftime("%H:%M:%S")}')
print('=' * 50)

for nb in notebooks:
    print(f'\n▶️  Ejecutando {nb}...')
    t0 = time.time()
    try:
        %run {nb}
        t = time.time() - t0
        resultados[nb] = ('✅', round(t/60, 1))
        print(f'✅  {nb} completado en {t/60:.1f} min')
    except Exception as e:
        t = time.time() - t0
        resultados[nb] = ('❌', round(t/60, 1))
        print(f'❌  {nb} falló en {t/60:.1f} min: {e}')

print('\n' + '=' * 50)
print('RESUMEN FINAL')
print('=' * 50)
for nb, (status, mins) in resultados.items():
    print(f'  {status}  {nb}  ({mins} min)')
print(f'\n🏁 Fin: {datetime.now().strftime("%H:%M:%S")}')
```
