```python
from pathlib import Path

ROOT = Path.cwd()
while not (ROOT / "src").exists():
    ROOT = ROOT.parent

print(f"Estructura de {ROOT / 'src'}:\n")
for item in sorted((ROOT / "src").rglob("*.py")):
    # Saltar __pycache__ y archivos ocultos
    if "__pycache__" in str(item) or item.name.startswith("."):
        continue
    rel = item.relative_to(ROOT / "src")
    size_kb = item.stat().st_size / 1024
    print(f"  src/{rel}  ({size_kb:.1f} KB)")
```
