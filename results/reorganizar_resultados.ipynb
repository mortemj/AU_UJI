````python
# 1. CONFIG
import sys
import shutil
from pathlib import Path

def _encontrar_root(start: Path) -> Path:
    for parent in [start] + list(start.parents):
        if (parent / 'src').is_dir():
            return parent
    raise FileNotFoundError('No se encontró src/')

ROOT = _encontrar_root(Path.cwd())

ORIGEN  = ROOT / 'data' / '05_modelado' / 'results'
DESTINO = ROOT / 'results' / 'fase5'

print(f'ROOT:    {ROOT}')
print(f'ORIGEN:  {ORIGEN}')
print(f'DESTINO: {DESTINO}')
print(f'Existe ORIGEN: {ORIGEN.exists()}')
````

````python
# 2. VERIFICAR QUÉ HAY EN ORIGEN
ficheros = sorted(ORIGEN.glob('*'))
print(f'Ficheros en {ORIGEN.name}:')
for f in ficheros:
    size_kb = f.stat().st_size / 1024
    print(f'  {f.name:50s} {size_kb:8.1f} KB')
print(f'\nTotal: {len(ficheros)} ficheros')
````

````python
# 3. COPIAR A results/fase5/
# Solo copia — NO borra nada del origen.

DESTINO.mkdir(parents=True, exist_ok=True)

copiados = []
errores  = []

for f in ORIGEN.glob('*'):
    if f.is_file():
        dest_f = DESTINO / f.name
        try:
            shutil.copy2(f, dest_f)
            copiados.append(f.name)
            print(f'  ✅ {f.name}')
        except Exception as e:
            errores.append((f.name, str(e)))
            print(f'  ❌ {f.name}: {e}')

print(f'\nCopiados: {len(copiados)} | Errores: {len(errores)}')
````

````python
# 4. VERIFICAR DESTINO
ficheros_dest = sorted(DESTINO.glob('*'))
print(f'Ficheros en results/fase5/:')
for f in ficheros_dest:
    size_kb = f.stat().st_size / 1024
    print(f'  {f.name:50s} {size_kb:8.1f} KB')
print(f'\nTotal: {len(ficheros_dest)} ficheros')
print('\n✅ Copia completada. Origen intacto.')
print('Siguiente paso: verificar que Fase 6 funciona con nuevas rutas.')
````
