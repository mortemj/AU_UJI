```python
# ============================================================================
# VERIFICACIÓN DE PARQUETS - AutoML y EDA
# ============================================================================

import sys
import pandas as pd
from pathlib import Path

ROOT = Path.cwd()
for _ in range(5):
    if (ROOT / 'src').exists(): break
    ROOT = ROOT.parent

print(f'ROOT: {ROOT}')
print()

# Archivos esperados
archivos = [
    ('AutoML (final)', ROOT / 'data' / 'automl' / 'df_exp_automl_target_final.parquet'),
    ('AutoML (intermedio)', ROOT / 'data' / '03_features' / 'df_exp_automl_target.parquet'),
    ('EDA (final)', ROOT / 'data' / '04_eda' / 'df_exp_target_eda_final.parquet'),
    ('EDA (intermedio)', ROOT / 'data' / '03_features' / 'df_exp_target_eda.parquet'),
    ('Features base', ROOT / 'data' / '03_features' / 'df_expediente_features.parquet'),
    ('Expediente base', ROOT / 'data' / '03_features' / 'df_expediente_base.parquet'),
    ('Alumno limpio', ROOT / 'data' / '03_features' / 'df_alumno_limpio.parquet'),
]

print('VERIFICACIÓN DE PARQUETS')
print('=' * 70)

for nombre, ruta in archivos:
    if ruta.exists():
        df = pd.read_parquet(ruta)
        tiene_target = '✅ abandono' if 'abandono' in df.columns else ''
        tipos_obj = df.select_dtypes(include=['object']).columns.tolist()
        texto = f'⚠️ TEXTO: {tipos_obj}' if tipos_obj else '✅ numérico'
        print(f'✅ {nombre:22s} | {df.shape[0]:>6,} × {df.shape[1]:>3} | {tiene_target} {texto}')
        del df
    else:
        print(f'❌ {nombre:22s} | {ruta.relative_to(ROOT)}')

# Results AutoML
print(f'\nRESULTADOS AUTOML')
print('-' * 70)
automl_dir = ROOT / 'data' / 'automl'
for f in sorted(automl_dir.glob('results_*.parquet')):
    df = pd.read_parquet(f)
    print(f'✅ {f.name:40s} | {len(df)} modelos')
    del df

if not list(automl_dir.glob('results_*.parquet')):
    print('  (ningún resultado todavía — ejecutar AutoML M01-M05)')
```
