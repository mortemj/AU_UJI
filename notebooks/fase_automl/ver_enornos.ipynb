```python
import json
from pathlib import Path

ROOT = Path.cwd()
for _ in range(6):
    if (ROOT / 'src').exists():
        break
    ROOT = ROOT.parent

DIR_AUTOML = ROOT / 'notebooks' / 'fase_automl'

notebooks = [
    'fautoml_m02_lazypredict.ipynb',
    'fautoml_m03_pycaret.ipynb',
    'fautoml_m04_h2o.ipynb',
    'fautoml_m05_autogluon.ipynb',
]

for nb_name in notebooks:
    ruta = DIR_AUTOML / nb_name
    try:
        nb = json.load(open(ruta, encoding='utf-8'))
        kernel = nb['metadata']['kernelspec']['display_name']
        print(f'{nb_name}: {kernel}')
    except FileNotFoundError:
        print(f'{nb_name}: NO ENCONTRADO en {ruta}')
```
