```python
import sys
from pathlib import Path

# Detectar ROOT (igual que todos tus notebooks)
ROOT = Path.cwd()
for _ in range(5):
    if (ROOT / 'src').exists():
        break
    ROOT = ROOT.parent

print(f'ROOT: {ROOT}')

# Buscar el parquet
import pandas as pd

posibles = [
    ROOT / 'data' / '05_automl' / 'df_exp_automl_target_final.parquet',
    ROOT / 'data' / '05_automl' / 'df_exp_automl_target.parquet',
    ROOT / 'data' / '04_features' / 'df_exp_automl_target.parquet',
    ROOT / 'data' / '04_features' / 'df_exp_automl_target_final.parquet',
    ROOT / 'data' / 'automl' / 'df_exp_automl_target_final.parquet',
]

print('\nBUSCANDO PARQUET AutoML...\n')
for p in posibles:
    if p.exists():
        df = pd.read_parquet(p)
        print(f'✅ ENCONTRADO: {p.relative_to(ROOT)}')
        print(f'   Shape: {df.shape}')
        print(f'   Columnas: {list(df.columns)}')
        tiene_target = 'abandono' in df.columns
        print(f'   Target "abandono": {"✅" if tiene_target else "❌"}')
        if tiene_target:
            print(f'   Abandono=1: {(df["abandono"]==1).sum()} ({(df["abandono"]==1).mean()*100:.1f}%)')
        # Tipos: ¿hay object/string? (problema para AutoML)
        tipos_obj = df.select_dtypes(include=['object']).columns.tolist()
        if tipos_obj:
            print(f'   ⚠️ Columnas tipo TEXTO: {tipos_obj}')
        else:
            print(f'   ✅ Todo numérico (listo para AutoML)')
        break
    else:
        print(f'❌ {p.relative_to(ROOT)}')
else:
    print('\n⚠️ NO SE ENCONTRÓ NINGÚN PARQUET')
    print('   Hay que ejecutar Fase 3 M04a + M05 para generarlo')
```
